# Importing the required libraries

```python
import numpy as np
import pandas as pd
import statistics
```

# Reading the dataset

```python
delivery=pd.read_csv('deliveries.csv')
delivery.head(5)
```

```python
#Determining the total numbers of rows and columns present in the dataset
delivery.shape
```

```python
#Reading the names of the columns
delivery.columns
```

# Finding out the batsmen with highest strike rate

```python
#batsman who has played more than 200 balls
a=delivery.groupby('batsman')['batsman_runs'].count()
b=delivery.groupby('batsman')['batsman_runs'].count()>200
c=a[b].index.tolist()
c
```

```python
#Finding out the strike rate
d=delivery[delivery['batsman'].isin(c)]
runs=d.groupby('batsman')['batsman_runs'].sum()
balls=d.groupby('batsman')['batsman_runs'].count()
strike_rate=[]
strike_rate=(runs/balls*100).sort_values(ascending=False)
```

```python
strike_rate.head()
```

```python
#Finding the mean
mean_batsman=statistics.mean(strike_rate)
mean_batsman
```

```python
#Finding the standard deviation
sd_batsman=statistics.stdev(strike_rate)
sd_batsman
```

# Finding out the  most economical bowlers

```python
delivery['dismissal_kind'].value_counts()
dismissal=['caught','bowled','lbw','stumped','caught and bowled','hit wicket']
out=delivery[delivery['dismissal_kind'].isin(dismissal)]
bowler=out.groupby('bowler')['dismissal_kind'].count().sort_values(ascending=False).index.tolist()
bowler
```

```python
#bowlers who have bowled more than 100 balls
i=delivery.groupby('bowler')['total_runs'].count()
j=delivery.groupby('bowler')['total_runs'].count()>100
k=i[j].index.tolist()
```

```python
#Calculating the runs given
new_bowler=delivery[delivery['bowler'].isin(k)]
runs_given=new_bowler.groupby('bowler')['total_runs'].sum().sort_values(ascending=False)
bowled_balls=new_bowler.groupby('bowler')['total_runs'].count().sort_values(ascending=False)
```

```python
#Finding out the overs bowled
bowled_overs=bowled_balls/6
```

```python
#Finding out the economy
economy=[]
economy=(runs_given/bowled_overs).sort_values(ascending=True)
economy
```

```python
#Finding out the mean
mean_bowler=statistics.mean((runs_given/bowled_overs).sort_values(ascending=False))
mean_bowler
```

```python
#Finding out the standard deviation
sd_bowler=statistics.stdev((runs_given/bowled_overs).sort_values(ascending=False))
sd_bowler
```

# Function defined to compare the batsman with bowler by calculating their Z-Scores

```python
def batsman_vs_bowler():
    print("Enter the name of the batsman")
    batsman_name=input()
    x=strike_rate[batsman_name]
    zscore_batsman=(x-mean_batsman)/sd_batsman #finding Z-score for the batsman

    print("Enter the name of the bowler")
    bowler_name=input()
    y=economy[bowler_name]
    zscore_bowler=(mean_bowler-y)/sd_bowler #Finding Z-Score for bowler
    
    if(zscore_batsman>zscore_bowler):   #comparing the two Z-Scores obtained and finnally displaying the result
        print("The better T20 player is "+batsman_name)
    else:
        print("The better T20 player is "+bowler_name)
```

# The final ouput showing the better T20 player

```python
batsman_vs_bowler()
```

```python
batsman_vs_bowler()
```
